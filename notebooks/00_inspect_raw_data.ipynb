```python
import sys
from pathlib import Path

# ถอยกลับไปที่ Root directory เพื่อให้เรียกไฟล์จาก src/ โฟลเดอร์ได้
project_root = Path.cwd().parent

if str(project_root) not in sys.path:
    sys.path.append(str(project_root))

from src.utils.raster_qc import run_qc, assert_qc_passed

raw_dir = project_root / "data/raw/rain_obs_monthly"

res = run_qc(
    folder=raw_dir,
    pattern="o_th*.tif",
    baseline_month=None,
    do_quick_stats=True,
)

print("--- สรุปช่วงเวลา ---")
display(res["df_range"].head())

print("\n--- ความผิดปกติของ Metadata (เทียบกับ Baseline) ---")
if res["df_incons"].empty:
    print("✅ Metadata ทุกไฟล์ตรงตามมาตรฐาน")
else:
    display(res["df_incons"])

print("\n--- สถิติข้อมูล (หาค่าติดลบ/Min/Max) ---")
display(res["df_stats"].head())

# hard gate (ตั้ง allow_missing_months ตามที่คุณต้องการ)
assert_qc_passed(
    df_incons=res["df_incons"],
    df_selection_anom=res["df_selection_anom"],
    df_missing=res["df_missing"],
    allow_missing_months=False,
)
```
